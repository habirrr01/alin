# **TRANSFORMASI LINIER**

# *Definisi Transformasi Linier*
Transformasi linear adalah sebuah fungsi yang memetakan suatu ruang vektor ke ruang vektor lainnya sedemikian rupa sehingga operasi penjumlahan vektor dan perkalian vektor dengan skalar tetap berlaku. Dalam konteks ini, jika $ T $ adalah sebuah transformasi linier, maka untuk setiap vektor $ \mathbf{u} $ dan $ \mathbf{v} $ serta skalar $ \alpha $ berlaku:

1. **Penjumlahan**:
   $
   T(\mathbf{u} + \mathbf{v}) = T(\mathbf{u}) + T(\mathbf{v})
   $

2. **Perkalian Skalar**:
   $
   T(\alpha \mathbf{u}) = \alpha T(\mathbf{u})
   $

# *Contoh Sifat Penjumlahan*

Misalkan $T: \mathbb{R}^2 \rightarrow \mathbb{R}^2$ adalah transformasi linier yang didefinisikan oleh matriks
$
T\binom{x}{y}=\left(\begin{array}{ll}
2 & 3 \\
1 & 4
\end{array}\right)\binom{x}{y}
$
dengan vektor-vektor $\mathbf{u}=\binom{1}{2}$ dan $\mathbf{v}=\binom{3}{4}$.
1. Hitung $T(\mathbf{u})$ dan $T(\mathbf{v})$ :
$
\begin{gathered}
T(\mathbf{u})=\left(\begin{array}{ll}
2 & 3 \\
1 & 4
\end{array}\right)\binom{1}{2}=\binom{2 \cdot 1+3 \cdot 2}{1 \cdot 1+4 \cdot 2}=\binom{2+6}{1+8}=\binom{8}{9} \\
T(\mathbf{v})=\left(\begin{array}{ll}
2 & 3 \\
1 & 4
\end{array}\right)\binom{3}{4}=\binom{2 \cdot 3+3 \cdot 4}{1 \cdot 3+4 \cdot 4}=\binom{6+12}{3+16}=\binom{18}{19}
\end{gathered}
$
2. Hitung $T(\mathbf{u}+\mathbf{v})$ :
$
\begin{gathered}
\mathbf{u}+\mathbf{v}=\binom{1}{2}+\binom{3}{4}=\binom{4}{6} \\
T(\mathbf{u}+\mathbf{v})=T\binom{4}{6}=\left(\begin{array}{ll}
2 & 3 \\
1 & 4
\end{array}\right)\binom{4}{6}=\binom{2 \cdot 4+3 \cdot 6}{1 \cdot 4+4 \cdot 6}=\binom{8+18}{4+24}=\binom{26}{28}
\end{gathered}
$
3. Hitung $T(\mathbf{u})+T(\mathbf{v})$ :
$
T(\mathbf{u})+T(\mathbf{v})=\binom{8}{9}+\binom{18}{19}=\binom{8+18}{9+19}=\binom{26}{28}
$
4. Bandingkan hasil-hasil tersebut:
$
\begin{gathered}
T(\mathbf{u}+\mathbf{v})=\binom{26}{28} \\
T(\mathbf{u})+T(\mathbf{v})=\binom{26}{28}
\end{gathered}
$

 Dari sini kita lihat bahwa
$
T(\mathbf{u}+\mathbf{v})=T(\mathbf{u})+T(\mathbf{v})
$

### Contoh Sifat Perkalian Skalar

Misalkan $T$ adalah transformasi linier yang sama seperti di atas, dan kita gunakan vektor $\mathbf{w}=\binom{2}{3}$ serta skalar $\alpha=5$.
1. Hitung $T(\alpha \mathbf{w})$ :
$
\begin{gathered}
\alpha \mathbf{w}=5\binom{2}{3}=\binom{5 \cdot 2}{5 \cdot 3}=\binom{10}{15} \\
T(\alpha \mathbf{w})=T\binom{10}{15}=\left(\begin{array}{ll}
2 & 3 \\
1 & 4
\end{array}\right)\binom{10}{15}=\binom{2 \cdot 10+3 \cdot 15}{1 \cdot 10+4 \cdot 15}=\binom{20+45}{10+60}=\binom{65}{70}
\end{gathered}
$
2. Hitung $\alpha T(\mathbf{w})$ :
$
\begin{gathered}
T(\mathbf{w})=\left(\begin{array}{ll}
2 & 3 \\
1 & 4
\end{array}\right)\binom{2}{3}=\binom{2 \cdot 2+3 \cdot 3}{1 \cdot 2+4 \cdot 3}=\binom{4+9}{2+12}=\binom{13}{14} \\
\alpha T(\mathbf{w})=5\binom{13}{14}=\binom{5 \cdot 13}{5 \cdot 14}=\binom{65}{70}
\end{gathered}
$
3. Bandingkan hasil-hasil tersebut:
$
\begin{aligned}
& T(\alpha \mathbf{w})=\binom{65}{70} \\
& \alpha T(\mathbf{w})=\binom{65}{70}
\end{aligned}
$

 Dari sini kita lihat bahwa
$
T(\alpha \mathbf{w})=\alpha T(\mathbf{w})
$

### Mencari Matriks A dari Transformasi Linear

Misal mencari matriks $A$ dari transformasi linier $T$ yang diberikan oleh:
$
\begin{aligned}
& T\binom{2}{0}=\binom{-2}{0} \\
& T\binom{2}{2}=\binom{-2}{2}
\end{aligned}
$

Mari kita mulai dengan mengekspresikan basis standar $\mathbf{e}_1=\binom{1}{0}$ dan $\mathbf{e}_2=\binom{0}{1}$ dalam bentuk kombinasi linear dari vektor-vektor yang transformasinya diketahui, yaitu $\binom{2}{0}$ dan $\binom{2}{2}$.
1. Menentukan $T\left(\mathbf{e}_1\right)$ :
Basis standar:
$
\mathbf{e}_1=\binom{1}{0}
$

Kita ekspresikan $\binom{1}{0}$ sebagai kombinasi linear dari $\binom{2}{0}$ dan $\binom{2}{2}$ :
$
\binom{1}{0}=a\binom{2}{0}+b\binom{2}{2}
$

Dari sini, kita mendapatkan sistem persamaan:
$
\left\{\begin{array}{l}
2 a+2 b=1 \\
0 a+2 b=0
\end{array}\right.
$

Dari persamaan kedua:
$
2 b=0 \rightarrow b=0
$

Substitusi $b=0$ ke dalam persamaan pertama:
$
2 a+2(0)=1 \rightarrow 2 a=1 \rightarrow a=\frac{1}{2}
$

Jadi, kita punya:
$
\binom{1}{0}=\frac{1}{2}\binom{2}{0}+0\binom{2}{2}
$

Karena $T$ adalah linier:
$
\begin{aligned}
& \text { Karena } T\binom{1}{0}=\frac{1}{2} T\binom{2}{0}+0 T\binom{2}{2} \\
& T\binom{1}{0}=\frac{1}{2}\binom{-2}{0}+0\binom{-2}{2} \\
& T\binom{1}{0}=\binom{-1}{0}
\end{aligned}
$

Jadi, kolom pertama dari matriks $A$ adalah:
$
\binom{-1}{0}
$

2. Menentukan $T\left(\mathbf{e}_2\right)$ :
Basis standar:
$\mathbf{e}_2=\binom{0}{1}$
Kita ekspresikan $\binom{0}{1}$ sebagai kombinasi linear dari $\binom{2}{0}$ dan $\binom{2}{2}$ :
$
\binom{0}{1}=a\binom{2}{0}+b\binom{2}{2}
$

Dari sini, kita mendapatkan sistem persamaan:
$
\left\{\begin{array}{l}
2 a+2 b=0 \\
0 a+2 b=1
\end{array}\right.
$

Dari persamaan kedua:
$
2 b=1 \rightarrow b=\frac{1}{2}
$

Substitusi $b=\frac{1}{2}$ ke dalam persamaan pertama:
$
2 a+2\left(\frac{1}{2}\right)=0 \rightarrow 2 a+1=0 \rightarrow 2 a=-1 \rightarrow a=-\frac{1}{2}
$

Jadi, kita punya:
$
\binom{0}{1}=-\frac{1}{2}\binom{2}{0}+\frac{1}{2}\binom{2}{2}
$

Karena $T$ adalah linier:
$
\begin{aligned}
& T\binom{0}{1}=-\frac{1}{2} T\binom{2}{0}+\frac{1}{2} T\binom{2}{2} \\
& T\binom{0}{1}=-\frac{1}{2}\binom{-2}{0}+\frac{1}{2}\binom{-2}{2} \\
& T\binom{0}{1}=\binom{1}{0}+\binom{-1}{1} \\
& T\binom{0}{1}=\binom{0}{1}
\end{aligned}
$

Jadi, kolom kedua dari matriks $A$ adalah:
$
\binom{0}{1}
$

Maka matriks $A$ yang memenuhi transformasi linier $T$ adalah:
$
A=\left(\begin{array}{cc}
-1 & 0 \\
0 & 1
\end{array}\right)
$


# *Apa Itu Refleksi?*
Refleksi adalah sebuah transformasi yang memindahkan setiap titik suatu objek ke posisi cerminan terhadap garis tertentu, seperti sumbu $x$, sumbu $y$, atau garis lain seperti $y=x$.

# *Sifat-Sifat Refleksi*
1. Jarak ke Cermin: Jarak dari titik awal ke cermin sama dengan jarak dari cermin ke bayangan titik tersebut.
2. Garis Tegak Lurus: Garis yang menghubungkan titik awal dan bayangannya selalu tegak lurus terhadap cermin.

# *Refleksi Terhadap Sumbu Koordinat*
1. Refleksi terhadap sumbu $x$ :
- Sumbu $x$ sebagai cermin.
- Titik $(x, y)$ direfleksikan menjadi $(x,-y)$.
- Matriksnya: $\left(\begin{array}{cc}1 & 0 \\ 0 & -1\end{array}\right)$
2. Refleksi terhadap sumbu $y$ :
- Sumbu $y$ sebagai cermin.
- Titik $(x, y)$ direfleksikan menjadi $(-x, y)$.
- Matriksnya: $\left(\begin{array}{cc}-1 & 0 \\ 0 & 1\end{array}\right)$

# *Refleksi Terhadap Garis*
1. Refleksi terhadap garis $y=x$ :
- Garis $y=x$ sebagai cermin.
- Titik $(x, y)$ direfleksikan menjadi $(y, x)$.
- Matriksnya: $\left(\begin{array}{ll}0 & 1 \\ 1 & 0\end{array}\right)$
2. Refleksi terhadap garis $y=-x$ :
- Garis $y=-x$ sebagai cermin.
- Titik $(x, y)$ direfleksikan menjadi $(-y,-x)$.
- Matriksnya: $\left(\begin{array}{cc}0 & -1 \\ -1 & 0\end{array}\right)$

Contoh
Misalkan kita punya transformasi linier:
$
\begin{aligned}
& T\binom{2}{0}=\left(\begin{array}{c}
-2 \\
0 \\
2 \\
2 \\
2
\end{array}\right)=\binom{2}{2}
\end{aligned}
$

Ini berarti bahwa komponen $x$ dari setiap titik berubah tanda, tetapi komponen $y$ tetap sama. Ini mirip dengan refleksi terhadap sumbu $y$.

Menentukan Matriks Transformasi
Kita ingin mencari matriks $A$ untuk transformasi ini.
1. Untuk titik $\binom{2}{0}$ :
- Setelah transformasi menjadi $\binom{-2}{0}$.
- Jika kita bagi dengan 2, kita dapatkan: $T\binom{1}{0}=\binom{-1}{0}$.
2. Untuk titik $\binom{2}{2}$ :
- Setelah transformasi menjadi $\binom{-2}{2}$.
- Jika kita bagi dengan 2, kita dapatkan: $T\binom{1}{1}=\binom{-1}{1}$.

Jadi, kita bisa simpulkan bahwa matriks transformasi $A$ adalah:
$
A=\left(\begin{array}{cc}
-1 & 0 \\
0 & 1
\end{array}\right)
$

Dalam contoh ini, transformasi yang diberikan adalah refleksi terhadap sumbu $y$, yang diwakili oleh matriks:
$
A=\left(\begin{array}{cc}
-1 & 0 \\
0 & 1
\end{array}\right)
$
